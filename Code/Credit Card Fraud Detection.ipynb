# Feature Engineering

```python
#Feature scaling is a preprocessing technique used in data science and machine learning to standardize the range of independent variables or features of the dataset. 
#The goal of feature scaling is to ensure that all features contribute equally to the computation of models, avoiding biases towards features with larger scales.
#Robust scaling is particularly useful when dealing with features that may have outliers or are not normally distributed. 
#Robust scaling is less sensitive to outliers compared to other scaling methods like standardization (Z-score normalization).
from sklearn.preprocessing import StandardScaler, RobustScaler
testdata['scaled_amount'] = RobustScaler().fit_transform(testdata['Amount'].values.reshape(-1, 1))
testdata['scaled_time'] = RobustScaler().fit_transform(testdata['Time'].values.reshape(-1, 1))

# Make a new dataset named "df_scaled" dropping out original "Time" and "Amount"
df_scaled = testdata.drop(['Time', 'Amount'], axis=1, inplace=False)
df_scaled.head()
```

```python
#As the data is highly unbalanced, we perform oversampling to add random values that give "1" under class (indicating the transaction was fraudulent)
def prep_data(testdata):
    X = testdata.drop(['Class'], axis=1, inplace=False)  
    X = np.array(X).astype(float)  # Use float instead of np.float
    y = testdata[['Class']]  
    y = np.array(y).astype(float)  # Use float instead of np.float
    return X, y

# Preparing our predictors(X) and target(y) variables.

# Create X and y from the prep_data function 
predictors, target = prep_data(df_scaled)
```

```python
from sklearn.model_selection import train_test_split
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import RandomOverSampler
from imblearn.over_sampling import SMOTE
from imblearn.over_sampling import BorderlineSMOTE
from imblearn.pipeline import Pipeline # Inorder to avoid testing model on sampled data
#imported the required libraries


# Create the training and testing sets
X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size=.3, random_state=0)

#We had to downgrade to scikit-learn version 1.2.2 in order for the current cell to run.
#The command used for the above was pip install scikit-learn==1.2.2
```

```python
undersam = RandomUnderSampler(random_state=0)
#Random under-sampling aims to balance the class distribution by randomly removing samples from the majority class
oversam = RandomOverSampler(random_state=0)
#Random over-sampling involves duplicating random samples from the minority class to balance the class distribution.
smote = SMOTE(sampling_strategy='auto', random_state=0)
#SMOTE (Synthetic Minority Over-sampling Technique)  generates synthetic samples for the minority class by considering the nearest neighbors.
borderlinesmote = BorderlineSMOTE(kind='borderline-2', random_state=0)
#BorderlineSMOTE is a variation of SMOTE that focuses on the samples that are close to the decision boundary (i.e., borderline samples).


# resample the training data
X_undersam, y_undersam = undersam.fit_resample(X_train, y_train)  
X_oversam, y_oversam = oversam.fit_resample(X_train, y_train) 
X_smote, y_smote = smote.fit_resample(X_train, y_train)  
X_borderlinesmote, y_borderlinesmote = borderlinesmote.fit_resample(X_train, y_train) 
```

# Model Training and testing

```python
from sklearn.linear_model import LogisticRegression
# Fit a logistic regression model to our data
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)

# Obtain model predictions
y_pred = log_reg.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
from sklearn.linear_model import LogisticRegression
# Fit a logistic regression model to our data
log_reg = LogisticRegression()
log_reg.fit(X_oversam, y_oversam)

# Obtain model predictions
y_pred = log_reg.predict(X_test)
accuracy2 = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy2)
```

```python
from sklearn.linear_model import LogisticRegression
# Fit a logistic regression model to our data
log_reg = LogisticRegression()
log_reg.fit(X_smote, y_smote)

# Obtain model predictions
y_pred = log_reg.predict(X_test)
accuracy3 = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy3)
```
